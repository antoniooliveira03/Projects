```python
import pandas as pd
```

```python
transaction = pd.read_csv("C:/Users/aprataso/Downloads/final_data/transaction_data_enriched.csv")
prod = pd.read_csv("C:/Users/aprataso/Downloads/final_data/product_enriched.csv")
```

```python
merger = transaction.merge(prod, on = 'product_id', how = 'left')
```

```python
merger
```

```python
# Count the number of times each department was purchased per basket_id
counts = merger.groupby(['basket_id', 'department']).size().reset_index(name='count')
counts
```

```python
# Pivot the table to have departments as columns
pivot_counts = counts.pivot_table(index='basket_id', columns='department', values='count', fill_value=0)
```

```python
# Reset index to make basket_id a column again
pivot_counts = pivot_counts.reset_index()

# Merge with the original DataFrame
final_merged = pd.merge(transaction, pivot_counts, on='basket_id', how='left')
```

```python
final_merged = final_merged.groupby('basket_id')
```
