```python
import re 

hh['male'] = hh['hh_comp_desc'].str.contains(r'\bmale\b', flags=re.IGNORECASE).astype(int)
hh['female'] = hh['hh_comp_desc'].str.contains(r'\bfemale\b', flags=re.IGNORECASE).astype(int)
```

% of males and females ignoring observations where m = 0 and f = 0

```python
filtered_hh = hh[(hh['male'] != 0) | (hh['female'] != 0)]

# Calculate total counts
total_rows = len(filtered_hh)
male_count = filtered_hh['male'].sum()
female_count = filtered_hh['female'].sum()

# Calculate percentages
male_percentage = (male_count / total_rows) * 100
female_percentage = (female_count / total_rows) * 100

print(f'male {male_percentage}, female {female_percentage}')
```

assinging a gender based on those probabilities to observations where hh_comp_desc = 1 adult kids

```python
male_probability = male_count / total_rows
female_probability = female_count / total_rows

adult_kids_filter = hh['hh_comp_desc'] == '1 Adult Kids'

# Step 3: Assign gender based on probabilities, ensuring one gender is assigned
np.random.seed(0)  # For reproducibility

# Randomly assign 'male' or 'female' based on probabilities
assign_gender = np.random.choice(['male', 'female'], size=adult_kids_filter.sum(), p=[male_probability, female_probability])

# Update the DataFrame based on the random assignments
hh.loc[adult_kids_filter, 'male'] = (assign_gender == 'male').astype(int)
hh.loc[adult_kids_filter, 'female'] = (assign_gender == 'female').astype(int)

```

prices

```python
transaction['loyalty_eu'] = transaction['loyalty_card_price'] * 1.07
transaction['non_loyalty_eu'] = transaction['non_loyalty_card_price'] * 1.07

transaction['loyalty_eu'] = transaction['loyalty_eu'].round(2)
transaction['non_loyalty_eu'] = transaction['non_loyalty_eu'].round(2)
```

```python
transaction = transaction.drop(['days_', 'loyalty_card_price', 'non_loyalty_card_price', 'sales_value'], axis = 1)
```

```python
basket = transaction.groupby(['household_key', 'basket_id', 'day']).agg({
    'product_id': lambda x: list(x),
    'quantity': lambda x: x.tolist(),
    'store_id': 'first',
    'retail_disc': lambda x: list(x),
    'trans_time': 'first',
    'week_no': 'first',
    'coupon_disc': lambda x: list(x),
    'coupon_match_disc': lambda x: list(x),
    'sales_value_eu': lambda x: list(x),
    'loyalty_eu': lambda x: list(x),
    'non_loyalty_eu': lambda x: list(x),
    'transaction_date': lambda x: list(x)
}).reset_index()

basket.sort_values(by='day')
```

```python
def find_non_numeric_values(lst):
    non_numeric_values = [item for item in lst if not isinstance(item, (int, float))]
    return non_numeric_values

basket['quantity'].apply(find_non_numeric_values)
```
