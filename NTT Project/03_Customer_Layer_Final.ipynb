### 2.1 Feature Engineering

Renaming Gender Variable

```python
art['gender'] = art['gender(s)']
art = art.drop(['gender(s)'], axis = 1)

data['gender'] = data['gender(s)']
data = data.drop(['gender(s)'], axis = 1)
```

Transforming *has_kids* in *artifical_data* to binary 

```python
art['has_kids'] = art['has_kids'].replace({
    True: 1,
    False: 0,
})
```

Drop Unnamed Column

```python
data = data.drop(['Unnamed: 0'], axis =1)
data.head(1)
```

### 2.2 Missing Values

Checking for Missing Values

```python
data.isna().sum()
```

As expected, there are 1699 missing values. These will be inputed by joining data from the *artificial_data* dataset

```python
art_iterator = iter(art.itertuples(index=False))

# Fill missing values in `data`
for row_index, row in data.iterrows():
    # Check if the row has missing values
    if row.isna().any():
        # Get the corresponding fill values from `art`
        art_row = next(art_iterator)
        
        # Fill missing values
        for column in data.columns:
            if pd.isna(row[column]):
                data.at[row_index, column] = getattr(art_row, column)

data.head(3)
```

Verifying the success of the operation

```python
data.isna().sum()
```

**Correction of misscomputed values**

Rectify the age_group misscomputed by Gretal.AI in the *artificial data* dataset

```python
for column in data.columns:
    if data[column].nunique() < 20:
        print(data[column].unique())
        print('-----')
```

```python
def map_age_group(avg_age):
    if avg_age == 21.5:
        return 'young adult'
    elif avg_age == 29.5:
        return 'adult'
    elif avg_age in [39.5, 49.5]:
        return 'middle-aged'
    else:
        return 'senior'

# Apply the function to create the new column
data['age_group'] = data['avg_age'].apply(map_age_group)
```

Verify the success of the operation

```python
for column in data.columns:
    if data[column].nunique() < 20:
        print(data[column].unique())
        print('-----')
```
