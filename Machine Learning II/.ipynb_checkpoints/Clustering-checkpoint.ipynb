## Variable Standardization

Creating a DataFrame with only numeric variables

```python
info_numeric = info_treated.loc[:, ~info_treated.columns.isin(
    ['customer_name', 'customer_id', 'education', 'Basic', 'Bsc', 'Msc', 'Phd'])]
```

## Customer Segmentation

### K-Means

Using dispersion to get the optimal number of clusters to be used

```python
dispersion = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(scaled_info)
    dispersion.append(kmeans.inertia_)
```

Elbow Method

```python
plt.plot(range(1, 20), dispersion, marker='o', color='orange')
plt.xlabel('Number of clusters')
plt.ylabel('Dispersion (inertia)')
plt.xticks(np.arange(0, 21, step=2))  
plt.vlines(7, 5000, 600000, color='red')

plt.show()
```

Performing K-Means with k clusters, (where k = [7, 10]) 

```python
kmeans7 = KMeans(n_clusters=7, random_state=0).fit(scaled_info)
kmeans8 = KMeans(n_clusters=8, random_state=0).fit(scaled_info)
kmeans9 = KMeans(n_clusters=9, random_state=0).fit(scaled_info)
kmeans10 = KMeans(n_clusters=10, random_state=0).fit(scaled_info)

info_treated['cluster_kmeans7'] = kmeans7.predict(scaled_info)
info_treated['cluster_kmeans8'] = kmeans8.predict(scaled_info)
info_treated['cluster_kmeans9'] = kmeans9.predict(scaled_info)
info_treated['cluster_kmeans10'] = kmeans10.predict(scaled_info)
```

Plotting a bar chart to analyse the number of observations in each cluster

```python
info_treated.groupby(['cluster_kmeans7']).size().plot(kind='bar', color='orange')
plt.show()
```

Taking a closer look at the average values within each cluster

```python
info_treated.groupby(['cluster_kmeans7']).mean().T
```

### SOM

Training the Algorithm

```python
som_info = MiniSom(15,
                   15,
                   22,
                   sigma=0.5,
                   learning_rate=1,
                   neighborhood_function='gaussian',
                   random_seed=42)
```

```python
# Set the seed of Numpy just to be sure of replicability
np.random.seed(42)

num_iterations = 1000

q_errors = []
for i in range(1, num_iterations):
    som_info.train_batch(scaled_info, i)
    q_errors.append(som_info.quantization_error(scaled_info))

```

Plotting how well the neurons are at representing the data points.

```python
plt.plot(range(1, num_iterations), q_errors, color='orange')
plt.show()
```

```python
som_info.train(scaled_info, 600)
```

Distance Map

```python
plt.figure(figsize=(8, 8))
plt.pcolor(som_info.distance_map().T, cmap=plt.cm.Oranges)  # Distance map shows the topological relationships
plt.colorbar()
```

Checking the maximum weight in each node

```python
feature_names = info_numeric.columns
feature_colors = {}

for i, feature in enumerate(feature_names):
  random_color = colors.to_rgba(colors.rgb2hex(np.random.rand(3)))
  feature_colors[i] = random_color
```

```python
W = som_info.get_weights()

plt.figure(figsize=(8, 8))
for i in np.arange(W.shape[0]):
    for j in np.arange(W.shape[1]):
        feature = np.argmax(W[i, j , :])
        plt.plot([i+.5], [j+.5], c=feature_colors[feature],
                marker='o', markersize=24)

legend_elements = [Patch(facecolor=feature_colors[i],
                         edgecolor='w',
                         label=f) for i, f in enumerate(feature_names)]

plt.legend(handles=legend_elements,
           loc='center left',
           bbox_to_anchor=(1, .95))
        
plt.xlim([0, 15])
plt.ylim([0, 15])
plt.show()
```

```python
info_numeric['winner_node'] = (
    [som_info.winner(scaled_info[i]) for i in range(0, len(scaled_info))]
)
```

### UMAP

Here we just fit a UMAP to our data, to be able to use it in the next two sections

```python
umap_object = umap.UMAP(n_neighbors=15, min_dist=0.3, random_state=42)
```

```python
X_sample = info_numeric.sort_index()
```

```python
umap_embedding = umap_object.fit_transform(X_sample)
```

### UMAP into DB Scan

```python
clusterable_embedding = umap.UMAP(
    n_neighbors=30,
    min_dist=0.0,
    n_components=2,
    random_state=42,
).fit_transform(X_sample)
```

```python
plt.scatter(clusterable_embedding[:, 0], clusterable_embedding[:, 1],s=0.5)
```

```python
labels = DBSCAN(eps=1, min_samples=20).fit_predict(clusterable_embedding)
```

```python
info_treated['cluster_umap_dbscan'] = labels
```

```python
clustered = (labels >= 0) 
# Plotting points not belonging to any cluster in gray (outliers)
plt.scatter(clusterable_embedding[~clustered, 0],
            clusterable_embedding[~clustered, 1],
            color=(0.5, 0.5, 0.5),
            s=0.3,
            alpha=0.5)

# Plotting points belonging to each cluster
plt.scatter(clusterable_embedding[clustered, 0],
            clusterable_embedding[clustered, 1],
            c=labels[clustered],
            s=0.5,
            cmap='Spectral');
```
