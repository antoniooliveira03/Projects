<font color='#FE7845'><font><h2>3.1. VADER Sentiment Analyzer</h2></font>
[Top &#129033;](#0) <br>

To begin this process, `SentimentIntensityAnalyzer` is used. This is a VADER sentiment analysis tool that will enable the execution of sentiment analysis without training. 

```python
vader = SentimentIntensityAnalyzer()
```

The aforementioned tool is implemented through the `vader_sa` function, as defined in the subsequent code cell. This function takes a song and a boolean parameter, determining whether to return compounded scores or individual scores (negative, neutral, positive).

```python
def vader_sa(song, compound = True):

    # Get the polarity scores (negative, neutral and positive) for the song
    polarity_scores_ = vader.polarity_scores(song)
    
    # If you want to compound the scores into one final score
    if compound:
        polarity = polarity_scores_["compound"]

    # If you want the three scores
    else:
        # The three separated scores
        polarity = polarity_scores_

    return polarity
```

The application of this function is extended to the 'clean_lyrics_sentiment' column, containing the pre-processed lyrics. Consequently, two additional columns are generated: 'vader_polarity_compound,' where compounded scores are stored, and 'vader_polarity_scores,' where individual negative, neutral, and positive scores are recorded.

```python
data["vader_polarity_compound"] = data["clean_lyrics_sentiment"].apply(lambda song : vader_sa(str(song)))

data["vader_polarity_scores"] = data["clean_lyrics_sentiment"].apply(lambda song : vader_sa(str(song), compound = False))
```

<font color='#FE7845'><font><h2>3.3. SentiWordNet</h2></font>
[Top &#129033;](#0) <br>

To continue the in-depth sentiment analysis on these lyrics, `SentiWordNet` is employed.

To start the process, it is first needed to tokenize the processed lyrics. This is executed below, saving this tokenized processed lyrics in the column 'TokenizedColumn'. 

```python
data['clean_lyrics_sentiment'] = data['clean_lyrics_sentiment'].astype(str)
data['TokenizedColumn'] = data['clean_lyrics_sentiment'].apply(lambda x: word_tokenize(x))
```

It's also important to attach a tag to each word in order to know its part of speech, based on its definition and context. This is conducted in the cell below, using the function `pos_tag`.

```python
data['PosTagsColumn'] = data['TokenizedColumn'].apply(lambda x: pos_tag(x))
```

Now, we still need to convert POS tags from the NLTK format to the WordNet format. <br>

The code below defines a function, `nltk_to_wordnet_pos`, for converting NLTK POS tags to WordNet format. Applied to a DataFrame column ('PosTagsColumn'), it creates a new column ('WordNet_PosTagsColumn') with WordNet POS tags for each word.


```python
def nltk_to_wordnet_pos(nltk_pos):
    if nltk_pos.startswith('N'):
        return wn.NOUN
    elif nltk_pos.startswith('V'):
        return wn.VERB
    elif nltk_pos.startswith('R'):
        return wn.ADV
    elif nltk_pos.startswith('J'):
        return wn.ADJ
    else:
        return None

# Convert NLTK POS tags to WordNet POS tags
data['WordNet_PosTagsColumn'] = data['PosTagsColumn'].apply(
    lambda tags: [(word, nltk_to_wordnet_pos(pos)) for word, pos in tags if nltk_to_wordnet_pos(pos) is not None]
)
```


In the followingellck, the function `get_sentiment_score` is defined to compute compounded polarity scorespolarity_compound']]


```python
def get_sentiment_score(word, tag):
    try:
        synsets = list(swn.senti_synsets(word, pos=tag))
        if synsets:
            return synsets[0].pos_score() - synsets[0].neg_score()
        else:
            return 0
    except KeyError:
        return 0

# Calculate sentiment scores and interpret the overall sentiment
data['SentiWordNet_polarity_compound'] = data['WordNet_PosTagsColumn'].apply(
    lambda tags: sum(get_sentiment_score(word, tag) for word, tag in tags))

# Display the resulting DataFrame
data[['clean_lyrics_sentiment', 'SentiWordNet_polarity_compound']]
```

The analysis process seems to have yielded success. 

To further explore and gain insights, the `.describe` method is employed on the 'SentiWordNet_polarity_compound' column. This action follows the grouping of data by the *tag* attribute. To finalize the visualizations, the average Sentiment Scores for the top 3 artists in each genre are displayed below. 

```python
for tag in data['tag'].unique():
    tag_df = data[data['tag'] == tag]

    # Find the top 3 artists within each tag
    top_artists = tag_df['artist'].value_counts().head(3).index.tolist()

    # Filter the data for only the top 3 artists within the tag
    top_artists_df = tag_df[tag_df['artist'].isin(top_artists)]

    # Iterate over the top 3 artists within the tag
    for artist in top_artists:
        artist_df = top_artists_df[top_artists_df['artist'] == artist]

        # Group by 'artist' and calculate the average sentiment score for each method
        average_scores = artist_df.groupby('artist')[compare].mean().reset_index()

        # Plotting
        plt.figure(figsize=(12, 8))  # Increased figure size

        # Plot each method's average score
        for i, method in enumerate(compare):
            plt.bar(i, average_scores[method], label=method, alpha=0.7)

        # Adding labels and title
        plt.xlabel('Sentiment Method')
        plt.ylabel('Average Sentiment Score')
        plt.title(f'Average Sentiment Scores for \"{artist}\" in {tag} Genre')
        
        # Set custom x-axis labels
        method_labels = ['Vader', 'TextBlob', 'SentiWordNet', 'Afinn']
        plt.xticks(range(len(compare)), method_labels)
        plt.grid(True)

        # Show the plot
        plt.show()
```
